```python
import cv2
import os

def combine_images(image_list1, image_list2, output_dir):
    # İki fotoğraf listesini eşleştir
    for img1_path, img2_path in zip(image_list1, image_list2):
        # İlk fotoğrafı yükle
        img1 = cv2.imread(img1_path)
        # İkinci fotoğrafı yükle
        img2 = cv2.imread(img2_path)

        # İki fotoğrafı birleştir
        combined_img = cv2.vconcat([img1, img2])

        # Yeni fotoğrafı kaydet
        output_filename = os.path.basename(img1_path)[:-4] + '_combined.jpg'  # Yeni dosya adı
        output_path = os.path.join(output_dir, output_filename)
        cv2.imwrite(output_path, combined_img)
        print(f"{output_filename} kaydedildi.")

# İlk fotoğraf listesi dizini
image_list1_dir = '/home/muhlabws4/Turkish_Open_2024_Antalya/Konya/08_02_2024/3th_court/cam_2/3th_court_2_cam_08_02_head_sensor/mhi_bgk'
# İkinci fotoğraf listesi dizini
image_list2_dir = '/home/muhlabws4/Turkish_Open_2024_Antalya/Konya/08_02_2024/3th_court/cam_2/3th_court_2_cam_08_02_head_sensor/kare'
# Yeni fotoğrafların kaydedileceği dizin
output_directory = '/home/muhlabws4/Turkish_Open_2024_Antalya/Konya/08_02_2024/3th_court/cam_2/3th_court_2_cam_08_02_head_sensor/combined_images'

# Eğer çıktı dizini yoksa oluştur
if not os.path.exists(output_directory):
    os.makedirs(output_directory)

# İlk fotoğraf listesini al
image_list1 = [os.path.join(image_list1_dir, img) for img in os.listdir(image_list1_dir)]
image_list2 = [os.path.join(image_list2_dir, img) for img in os.listdir(image_list2_dir)]

# Fotoğrafları birleştir
combine_images(image_list1, image_list2, output_directory)
```

```python
import cv2
import os

def resize_image(image, width, height):
    return cv2.resize(image, (width, height))

def combine_images_with_mhi(frame_dir, mhi_dir, output_dir):
    # Tüm frame resimlerini al
    frame_images = sorted([f for f in os.listdir(frame_dir) if f.endswith('.jpg')])

    # Tüm MHI resimlerini al
    mhi_images = sorted([f for f in os.listdir(mhi_dir) if f.endswith('.png')])

    # Yeniden boyutlandırma için hedef boyutlar
    target_width = 1280
    target_height = 720

    # Her bir MHI resmi için işlem yap
    for mhi_image in mhi_images:
        # MHI dosya adından başlangıç ve bitiş kare numaralarını al
        start_frame = int(mhi_image.split('_')[1])
        end_frame = int(mhi_image.split('_to_')[1][:-4])

        # MHI görüntüsünü yükle ve yeniden boyutlandır
        mhi_img = cv2.imread(os.path.join(mhi_dir, mhi_image))
        mhi_img = resize_image(mhi_img, target_width, target_height)

        # İlgili aralıktaki frame resimlerini bul ve birleştir
        combined_img = None
        for frame_image in frame_images:
            frame_num = int(frame_image.split('_')[1].split('.')[0])
            if start_frame <= frame_num <= end_frame:
                frame_img = cv2.imread(os.path.join(frame_dir, frame_image))
                if frame_img is not None:
                    # Frame görüntüsünü yeniden boyutlandır
                    frame_img = resize_image(frame_img, target_width, target_height)
                    if combined_img is None:
                        combined_img = frame_img
                    else:
                        # Frame görüntüsünü MHI üzerine bindir
                        combined_img = cv2.addWeighted(frame_img, 0.5, combined_img, 0.5, 0)

        if combined_img is not None:
            # MHI ve frame resimlerini birleştir
            final_img = cv2.vconcat([mhi_img, combined_img])

            # Yeni resmi kaydet
            output_filename = f"{mhi_image[:-4]}_combined.jpg"
            output_path = os.path.join(output_dir, output_filename)
            cv2.imwrite(output_path, final_img)
            print(f"{output_filename} kaydedildi.")

# Frame resimlerinin bulunduğu dizin
frame_directory = '/home/muhlabws4/dosyalar/diger/3_may_frame'
# MHI resimlerinin bulunduğu dizin
mhi_directory = '/home/muhlabws4/dosyalar/diger/3_may_resim'
# Yeni resimlerin kaydedileceği dizin
output_directory = '/home/muhlabws4/dosyalar/diger/3_may_combined_images'

# Eğer çıktı dizini yoksa oluştur
if not os.path.exists(output_directory):
    os.makedirs(output_directory)

# Resimleri birleştir
combine_images_with_mhi(frame_directory, mhi_directory, output_directory)
```
